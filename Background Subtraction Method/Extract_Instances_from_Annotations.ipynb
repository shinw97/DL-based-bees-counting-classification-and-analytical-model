```
import cv2
from google.colab.patches import cv2_imshow
```

```
i = 0
j = 0
for img in imgs:
    img_raw = cv2.imread(img['filename'])
    
    for obj in img['object']:
        x_min, y_min, x_max, y_max = obj['xmin'], obj['ymin'], obj['xmax'], obj['ymax']
        cropped = img_raw.copy()[y_min:y_max + 1, x_min: x_max + 1]
        t = obj['name']
        if t == 'bee-A':
            count = i
            i += 1
        else:
            count = j
            j += 1
        cv2.imwrite('MARDI-dataset/cropped/' + t + '/' + t + "_" + str(count) +".jpg", cropped)
        print()
        print(img['filename'] +' - ' +  t + ' --- MARDI-dataset/cropped/' + t + '/' + t + "_" + str(count) +".jpg")
#         break
#         cv2_imshow(cropped)
#     break
```
